```python
from moviepy.editor import VideoFileClip

from PIL import Image
from glob import glob

import sys
import cv2
import numpy as np
import traceback

from alpr_lib import alpr
```

```python
file_name = 'FILE190811-124346F.MP4'
in_dir = 'test_video'
out_dir = 'frame_image/out'
in_file_path = in_dir+'/'+file_name
out_file_path = out_dir+'/'+file_name
start_time = 82 # unit :second
end_time = 105 # unit :second
```

```python
myclip = VideoFileClip(in_file_path).subclip(start_time,end_time)
```

```python
i = 0
   
Vout = cv2.VideoWriter('%s' % (out_file_path),cv2.VideoWriter_fourcc(*'DIVX'), myclip.fps, myclip.size)    
    
for frame in myclip.iter_frames():
    if (True): 
        print ('i = %s' % (i)) 
        img = cv2.cvtColor(np.asarray(frame), cv2.COLOR_RGB2BGR) 
        Idraw, carsPos, lpsPos, lp_strs, Tcars, Ccars, IDcars = alpr(img)
        cv2.imwrite('%s/output_%s.png' % (out_dir, i), Idraw)
        Vout.write(Idraw)
    i = i+1
    
Vout.release()    
```
